```python
import requests
```

```python
URL = 'https://app.rakuten.co.jp/services/api/Travel/KeywordHotelSearch/20170426'
APP_ID = '1021205108156603837'
```

```python
prams = {
    'applicationId' : APP_ID,
    'keyword' : '熊本'
}
```

```python
res = requests.get(URL, prams)
```

```python
res.status_code
```

```python
hoge = res.json()
hoge
```
